```python
import pandas as pd
import re
```

```python
df_meta = pd.read_csv('../../accession_lists/big_control_meta.txt', sep='\t')
```

```python
df_data = pd.read_csv('../data_files/big_control.csv', index_col=0)
```

```python
df_data.head()
```

```python
def get_srr_accession(df):
    
    pattern = re.compile(r"[DSE]RR[0-9]+")
    
    file = df['file_location']
    
    return re.search(pattern, file).group(0)
```

```python
df_data['srr_acc'] = df_data.apply(get_srr_accession, axis=1)
```

```python
df_data.set_index('srr_acc',inplace=True)
```

```python
df_data.head()
```

```python
df_meta.set_index('Run', inplace=True)
```

```python
df_meta.head()
```

```python
del df_meta.index.name
del df_data.index.name
```

```python
df_meta.head()
```

```python
master_df = df_meta.join(df_data)
```

```python
master_df.to_csv('big_control.meta.csv')
```
