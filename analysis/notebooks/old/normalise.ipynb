```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('aml_with_spot_number.csv', index_col=0)
```

```python
df.head()
```

```python
# Normalise by dividing reads by 


df['norm_alignment_count'] = df['alignment_count'] / (df['total_spots']/ 1000000)
df['norm_alpha_wt_count'] = df['alpha_wt_count'] / (df['total_spots']/ 1000000)
df['norm_alpha_dup_count'] = df['alpha_dup_count'] / (df['total_spots']/ 1000000)
df['norm_beta_count'] = df['beta_count'] / (df['total_spots']/ 1000000)
df['norm_alpha_read_covers_snps_count_exact'] = df['alpha_read_covers_snps_count_exact'] / (df['total_spots']/ 1000000)
df['norm_alpha_dup_read_covers_snps_count_exact'] = df['alpha_dup_read_covers_snps_count_exact'] / (df['total_spots']/ 1000000)
df['norm_beta_read_covers_snps_count_exact'] = df['beta_read_covers_snps_count_exact'] / (df['total_spots']/ 1000000)
```

```python
# make some plots

raw_normalised_alignments = [df['norm_alpha_wt_count'].sum(), df['norm_alpha_dup_count'].sum(), df['norm_beta_count'].sum()]

ind = np.arange(3)


fig, ax = plt.subplots(2,1, figsize=(20,10))

bar1 = ax[0].bar(ind, raw_normalised_alignments, 0.35)

ax[0].set_title('Sum of Normalised Alignment Counts (Unfiltered)')
ax[0].set_xticks(ind)
ax[0].set_xticklabels(('alpha_wt', 'alpha_dup', 'beta'))
ax[0].set_xlabel('Transcript')
ax[0].set_ylabel('Hit Count (RPM)')


exact_hits_at_locus = [df['norm_alpha_read_covers_snps_count_exact'].sum(),
                       df['norm_alpha_dup_read_covers_snps_count_exact'].sum(),
                       df['norm_beta_read_covers_snps_count_exact'].sum()]

bar2 = ax[1].bar(ind, exact_hits_at_locus, 0.35)
ax[1].set_title('Sum of Normalised Alignment Counts (Exact Hits at Locus)')
ax[1].set_xticks(ind)
ax[1].set_xticklabels(('alpha_wt', 'alpha_dup', 'beta'))
ax[1].set_xlabel('Transcript')
ax[1].set_ylabel('Hit Count (RPM)')
plt.tight_layout()
```
