# Proportion Measuement

- Calculate the proportion of runs which have a duplication

```python
import pandas as pd
```

```python
df = pd.read_csv('../data_files/aml.3snps.norm.csv', index_col=0)
```

```python
df.head()
```

```python
#has this run got any alpha dup in it?

df['has_dup'] = df.apply(lambda x: False if x['norm_alpha_dup_read_covers_snps_count_exact'] == 0 else True, axis=1)
```

```python
#Look at the runs with at least one hit in alpha wt or alpha dup

filtered =  df[(df['norm_alpha_read_covers_snps_count_exact'] != 0) | (df['norm_alpha_dup_read_covers_snps_count_exact'] != 0)]
```

```python
filtered.shape
```

```python
filtered.head()
```

```python
value_counts = df['has_dup'].value_counts()

false_count = value_counts[0]
true_count = value_counts[1]

percentage = true_count / (true_count+ false_count) *100

print ('The proportion of all runs with the dup variation is {percent}%'.format(percent=round(percentage,2)))
```

```python
value_counts = filtered['has_dup'].value_counts()

false_count = value_counts[0]
true_count = value_counts[1]

percentage = true_count / (true_count+ false_count) *100

print ('The proportion of the runs with at least one hit in either alpha_wt or alpha_dup is {percent}%'.format(percent=round(percentage,2)))
```
