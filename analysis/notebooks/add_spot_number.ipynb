```python
def get_srr_accession(df):
    
    pattern = re.compile(r"[SE]RR[0-9]+")
    
    file = df['file_location']
    
    return re.search(pattern, file).group(0)
```

```python
df['srr_acc'] = df.apply(get_srr_accession, axis=1)
```

```python
df.head()
```

```python
def get_spot_count(df, email):
    
    srr_acc = df['srr_acc']
    
    print (srr_acc)
    
    # get SRR id
    Entrez.email = email
    
    error_count =0
    
    while error_count < 3:
        
        try:
    
            handle = Entrez.esearch(db="sra",term=srr_acc)
    
            record = Entrez.read(handle)
        
            handle.close()
    
            srr_id = record["IdList"][0]
        
            break
            
        except:
            
            print ('id', srr_acc, error_count)
            
            time.sleep(10)
            
            error_count = error_count +1
            
            
    # get SRR summary
    
    error_count =0
    
    while error_count < 3:
        
        try:  
    
            handle = Entrez.esummary(db='sra', id=srr_id)
    
            record = Entrez.read(handle)
    
            handle.close()
    
            my_xml = record[0]['Runs']
    
            # Parse XML
            xml_object = ET.fromstringlist(["<root>", my_xml, "</root>"])
    
            # Get total_spots (reads)
            for child in xml_object:
        
                if child.attrib['acc'] == srr_acc:
                
                    print (srr_acc, child.attrib['total_spots'])
            
                    return str(child.attrib['total_spots'])
            
            
            
            
        except:
            
            print ('summary', srr_acc, error_count)
            
            time.sleep(10)
            
            error_count = error_count +1
```

```python
df['total_spots'] = df.apply(get_spot_count, axis=1, args=['halsteadjs@nih.gov'])
```

```python
df['total_spots'] = df['total_spots'].astype(int)
```

```python
df['norm_alpha_read_covers_snps_count_exact'] = int((df['total_spots']/ 1000000))
```
