```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('data.3snps.csv', index_col=0)
```

```python
df.head()
```

```python
map_df = pd.read_csv('acc-to_spot_map.csv', index_col=0)
```

```python
map_df.head()
```

```python
df = df.join(map_df)
```

```python
df['norm_alpha_read_covers_snps_count_exact'] = df['alpha_read_covers_snps_count_exact'] / (df['total_spots']/ 1000000)
df['norm_alpha_dup_read_covers_snps_count_exact'] = df['alpha_dup_read_covers_snps_count_exact'] / (df['total_spots']/ 1000000)
df['norm_beta_read_covers_snps_count_exact'] = df['beta_read_covers_snps_count_exact'] / (df['total_spots']/ 1000000)
```

```python
df.head()
```

```python
filtered_df =df[(df['norm_alpha_read_covers_snps_count_exact'] != 0) | (df['norm_alpha_dup_read_covers_snps_count_exact'] != 0) ]
```

```python
filtered_df = filtered_df[['norm_alpha_read_covers_snps_count_exact',
                          'norm_alpha_dup_read_covers_snps_count_exact',
                          'norm_beta_read_covers_snps_count_exact']]
```

```python
filtered_df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,10))


df.plot(x='norm_alpha_read_covers_snps_count_exact',
        y='norm_alpha_dup_read_covers_snps_count_exact',
        kind='scatter',
        ax=ax)

ax.set_xlabel('alpha wt (RPM)')
ax.set_ylabel('alpha dup (RPM)')
ax.set_title('Tryptase Variant Expression in AML RNA Seq')
```
