```python
import pandas as pd
import glob
import numpy as np
```

```python
files = glob.glob('data/GSE67039_RAW/*')
```

```python
def create_master_frame(files):
    
    for i, file in enumerate(files):
        
        if i == 0:
            
            sample_name = '_'.join(file.split('/')[len(file.split('/'))-1].split('_')[0:2])
            
            print (sample_name)
                
            df = pd.read_csv(file,sep='\t', usecols=[0,3,5], header=None, index_col=[0,1], names=['chr', 'gene', sample_name] )

                          
        else:

            sample_name = '_'.join(file.split('/')[len(file.split('/'))-1].split('_')[0:2])
            
            next_df = pd.read_csv(file,sep='\t', usecols=[0,3,5], header=None, index_col=[0,1], names=['chr', 'gene', sample_name])
            df = df.merge(next_df, how='left', left_index=True, right_index=True)
            
            
                        
    return df
```

```python
df = create_master_frame(files)
```

```python
df = df.reset_index()

df.index = df['chr'] + "-" + df['gene']

#df.drop(['chr', 'gene'], inplace=True, axis=1)
```

```python
df = df[(df['chr'] != 'chrX') & (df['chr'] != 'chrY') ]
```
